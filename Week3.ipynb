```
import cv2 as cv 
from google.colab.patches import cv2_imshow 
from skimage import io 
import matplotlib.pyplot as plt 
import numpy as np 

img = cv.imread('/content/drive/MyDrive/images/cat.jpg')
image = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(image)
```

```
# Initialize values
print('-------------------------')
try:
    brightness = int(input('* Input nilai kecerahan [0-255]: '))
    contrast = float(input('* Input nilai kontras [1.0-3.0]: '))
except ValueError:
    print('Inputan hanya berupa angka')

original = cv.imread('/content/drive/MyDrive/images/cat.jpg')
new_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      new_image[y,x,c] = np.clip(contrast*original[y,x,c] + brightness, 0, 255)


final_frame = cv.hconcat((original, new_image))
cv2_imshow(final_frame)
```

```
print(' mengubah tingkat kecerahan citra dengan transform log ')
print(' -------------------------------- ')

try:
  brightness = int(input('Masukan nilai kecerahan: '))
except ValueError:
  print('Error, bukan angka')

original = image
brightness_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(brightness*(np.log(1.0+original[y,x,c])), 0, 255)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(cv.cvtColor(final_frame,cv.COLOR_BGR2RGB))
```

```
import cv2 as cv 
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
import math

img = cv.imread('/content/drive/MyDrive/images/cat.jpg')
img_color = cv.cvtColor(img,cv.COLOR_BGR2RGB)

plt.imshow(img_color)

# ekstraksi 3 channel warna
# image[x,y,c]
R, G, B = img_color[:,:,0], img_color[:,:,1],img_color[:,:,2]

gray_img = ( R + G + B) * 0.3

f, axarr = plt.subplots(1,2) #membuat 2 plot untuk 2 citra input dan output
axarr[0].imshow(img_color)
axarr[1].imshow(gray_img, cmap='gray')
```

```
# ekstraksi 3 channel warna
# image[x,y,c]
R, G, B = img_color[:,:,0], img_color[:,:,1],img_color[:,:,2]
max_channels = np.maximum(np.maximum(R, G), B) #np.max(color_img, axis=2)
min_channels = np.minimum(np.minimum(R, G), B) #np.min(color_img, axis=2)

gray_img = 0.5 * (max_channels + min_channels)

f, axarr = plt.subplots(1,2) #membuat 2 plot untuk 2 citra input dan output
axarr[0].imshow(img_color)
axarr[1].imshow(gray_img, cmap='gray')
```

```
R, G, B = img_color[:,:,0], img_color[:,:,1],img_color[:,:,2]

gray_img = (0.21 * R) + (0.72 * G) + (0.07 * B)

f, axarr = plt.subplots(1,2)
axarr[0].imshow(img_color)
axarr[1].imshow(gray_img, cmap='gray')
```
